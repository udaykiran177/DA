```python
import pandas as pd
```

```python
df=pd.read_csv('Z:\\DA\\files\\movies.csv')
df=df.fillna(0)


df
```

```python
# 1. (i). Find out the no of movies released in every month of the year 1995.

df.drop(df[df['Release Date']=="TBD"].index,axis=0,inplace=True)

df['Release Date']=df['Release Date'].apply(lambda x:pd.to_datetime(x))
df
```

```python
df[df['Release Date'].apply(lambda x:x.year)==1995].groupby([pd.Grouper(key='Release Date',axis=0,freq='m')])['Title'].count()
```

```python
# (ii). Find out the no. of movies released in every year from 1990 to 1998.
df[df['Release Date'].apply(lambda x:x.year).between(1990,1998)].groupby(pd.Grouper(key='Release Date',axis=0,freq='y'))['Title'].count()
```

```python
# 2.	(i). Find no. of movies released under each genre given in the database.

df.groupby(['Major Genre'])['Title'].count()
```

```python
# (ii). Find the movies under each genre with 1MDB rating >7 and rotten tomatoes rating > 60.

df[(df['IMDB Rating']>7) & (df['Rotten Tomatoes Rating']>60)].groupby('Major Genre')['Title'].count()
```

```python
# 3.	(i). Find the movies released under each fiction with each director in the ascending order of release dates.

m=df.groupby(['Director',df[df['Creative Type'].apply(lambda x:'Fiction' in str(x))]['Creative Type']]).first().sort_values('Release Date')
m
```

```python
# (ii). Find movies released under each distributor in the order of genre and director.
# pd.set_option('display.max_rows',100)

df.groupby(['Distributor','Major Genre','Director'])[].first()
```

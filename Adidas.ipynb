```python
#5.	Find the total sales of every retailer.


m.groupby(['Retailer'])['Total Sales'].sum()
```

```python
#6.	Find the total sales of the retailers in every city along with profit.

m.groupby(['Retailer','City'])[['Total Sales','Operating Profit']].sum()
```

```python
#7.	Find the total sales &profit of each product sold by the retailer.

m.groupby(['Retailer','Product'])[['Total Sales','Operating Profit']].sum()
```

```python
#8.	Find the units sold, total sales &profit of the products sold between the dates 1/1/2020 and 4/15/2020.

m[m['Invoice Date'].between('2020/01/01','2020/04/15')][['Units Sold','Total Sales','Operating Profit','Invoice Date']]
```

```python
#9.	Find the no of units sold of each product by each retailer in every city.


m.groupby(["Retailer",'City','Product'])['Units Sold'].first()
```

```python
#10.	Find the products with different price per unit in different cities with proper information.

m.groupby(['Product','City'])['Price per Unit'].first()
```

```python
#  11.	Find the total sales &profits of all products in every month.

m['Month']=[i.month for i in m['Invoice Date']]
m.groupby(['Month','Product'])[['Total Sales',"Operating Profit"]].sum()
```

```python
#  12.	Find the total sales &profit of the products in different sales methods in each city.

m.groupby(['City','Sales Method'])[['Total Sales','Operating Profit']].first()
```

```python
#13.	Find the retailers who sold the same product with different prices in different cities.

# pd.set_option('display.max_rows',None) #to display entire dataframe
m.groupby(['Retailer','Product','City'])['Price per Unit'].unique()
```

```python
# m.groupby(['Retailer','City'])['Price per Unit'].count()
#14.	Find the products whose sales raises in every month.


m.groupby(['Product',pd.Grouper(key="Invoice Date",axis=0,freq="M")])[['Total Sales']].first().diff().apply(lambda x:x<0).all(axis=1)
```

```python
#15.	Find the retailers whose profit increased every month.
m.groupby(['Retailer',pd.Grouper(key="Invoice Date",axis=0,freq="Y")])[['Operating Profit']].sum().diff().apply(lambda x:x>0).all()
```
